***Machine Learning Project to predict the purchase of iPhone by a person based on given features with the help of Naive Bayes Classifier***

**Bayes theorem :**Bayes' Theorem states that the conditional probability of an event, based on the occurrence of another event, is equal to the likelihood of the second event given the first event multiplied by the probability of the first event.

Screenshot 2023-01-10 220308.png

In the above Bayes formula P(A|B) denotes the **posterior probability** , P(B|A) denotes the **likelihood** , P(A) denotes the **prior probability** , P(B) denotes the **marginal likelihood.**

The objective is to predict if a customer will purchase iPhone or not based on given features in dataset... 

```python
#Import the dataset

import pandas as pd
import numpy as np
idata = pd.read_csv("iphone_purchase_records.csv")
X = idata.iloc[:,:-1].values
y = idata.iloc[:,3].values

#In the given the "Gender" attribute exists as text so we convert that into numerical form
```

```python
#In order to convert the Gender attribute to numerical form we import the LabelEncoder

from sklearn.preprocessing import LabelEncoder
labelEncoder_gender = LabelEncoder()
X[:,0] = labelEncoder_gender.fit_transform(X[:,0])
X = np.vstack(X[:,:]).astype(np.float)

#Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)

from sklearn.preprocessing import StandardScaler
ss_X = StandardScaler()
X_train = ss_X.fit_transform(X_train)
X_test = ss_X.transform(X_test)

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)

#Predict

from sklearn import metrics
y_pred = classifier.predict(X_test)

cm = metrics.confusion_matrix(y_test,y_pred)
print(cm)
accuracy = metrics.accuracy_score(y_test,y_pred)
print("Accuracy is:",accuracy)
precision = metrics.precision_score(y_test,y_pred)
print("Precision is:",precision)
recall = metrics.recall_score(y_test,y_pred)
print("Recall value:",recall)
```
